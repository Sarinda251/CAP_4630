# Problem 1

### Imports

```
# Python imports
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
```

## Function 1

**$\\f_1 = |x| + |y|, \:\: p_1 = (0,0,0)$**


### 3d plot

```
# Plot creation
ax = plt.axes(projection='3d')

# Function definition
def f1(x, y):
    return np.sqrt((x ** 2 + y ** 2))
X1, Y1 = np.meshgrid(np.linspace(-5,5,50), np.linspace(-5,5,50))
Z1 = f1(X1, Y1)

# Plot f1
ax.contour3D(X1, Y1, Z1, 100)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(10, 45)
ax.set_zlim(-1, 7)

# Plot p1
ax.plot([0.], [0.], [0.], markerfacecolor='purple', markeredgecolor='purple', marker='.', markersize=10)
ax.text(2, 2.5, 0, "p1 = (0,0,0)", color='purple', fontsize=12, fontweight='bold')

plt.show()
```

### Density plot

```
# Plot density
plt.contourf(X1, Y1, Z1, 20, cmap='RdGy')
plt.plot([0.], [0.], markerfacecolor='black', markeredgecolor='black', marker='.', markersize=10)
plt.text(-1,.5, "p1 = (0,0,0)", color='black', fontsize=12, fontweight='bold')
plt.colorbar();
```

### Contour plot

```
# Plot contour
plt.contour(X1, Y1, Z1, 20, cmap='RdGy')
plt.plot([0.], [0.], markerfacecolor='blue', markeredgecolor='blue', marker='.', markersize=10)
plt.text(0,.5, "p1 = (0,0,0)", color='blue', fontsize=12, fontweight='bold')
plt.show()
```

## Function 2

$\\f_2 = 1 - x^2 - y^2, \:\: p_2 = (0,0,1)$

### 3d plot

```
# Plot creation
ax = plt.axes(projection='3d')

# Function definition
def f2(x, y):
    return 1 - (x ** 2 + y ** 2)
X2, Y2 = np.meshgrid(np.linspace(-1,1,25), np.linspace(-1,1,25))
Z2 = f2(X2, Y2)

# Plot f2
ax.contour3D(X2, Y2, Z2, 20)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
ax.set_zlim(-1, 2)
ax.set_xticks(np.linspace(-1,1,5))
ax.set_yticks(np.linspace(-1,1,5))

# Plot p2
ax.plot([0.], [0.], [1.], markerfacecolor='purple', markeredgecolor='purple', marker='.', markersize=13)
ax.text(0, 0, 1.25, "p2 = (0,0,1)", color='purple', fontsize=12, fontweight='bold')

plt.show()
```

### Density plot

```
plt.contourf(X2, Y2, Z2, 20, cmap='RdGy')
plt.plot([0.], [0.], markerfacecolor='white', markeredgecolor='white', marker='.', markersize=10)
plt.text(-.25,.1, "p2 = (0,0,1)", color='white', fontsize=12, fontweight='bold')
plt.colorbar();
```

### Contour plot

```
# Plot contour
plt.contour(X2, Y2, Z2, 20, cmap='RdGy')
plt.plot([0.], [0.], markerfacecolor='black', markeredgecolor='black', marker='.', markersize=10)
plt.text(-.24,.1, "p2 = (0,0,1)", color='black', fontsize=12, fontweight='bold')
plt.show()
```

## Function 3

$\\f_3 = x^2 - y^2, \:\: 
p_3 = (0,0,0) \\
\nabla f_3(x,y) = \begin{bmatrix}2x\\2y\end{bmatrix}, \:\: \nabla f_3(0,0) = \begin{bmatrix}0\\0\end{bmatrix}$

### 3d plot

```
# Plot creation
ax = plt.axes(projection='3d')

# Function definition
def f3(x, y):
    return (x ** 2 - y ** 2)
X3, Y3 = np.meshgrid(np.linspace(-1,1,25), np.linspace(-1,1,25))
Z3 = f3(X3, Y3)

# Plot f3
ax.contour3D(X3, Y3, Z3, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(30, 60)
ax.set_zlim(-1, 2)
ax.set_xticks(np.linspace(-1,1,5))
ax.set_yticks(np.linspace(-1,1,5))

# Plot p3
ax.plot([0.], [0.], [0.], markerfacecolor='purple', markeredgecolor='purple', marker='.', markersize=13)
ax.text(1, 0, 1.1, "p3 = (0,0,0)", color='purple', fontsize=12, fontweight='bold')

plt.show()
```

### Density plot

```
plt.contourf(X3, Y3, Z3, 20, cmap='RdGy')
plt.plot([0.], [0.], markerfacecolor='black', markeredgecolor='black', marker='.', markersize=10)
plt.text(-.27,.1, "p3 = (0,0,0)", color='black', fontsize=12, fontweight='bold')
plt.colorbar();
```
